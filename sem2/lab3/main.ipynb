Задание 1:
Определите количество признаков и их типы, проверьте наличие пропущенных данных, выведите основные статистические характеристики (среднее значение, стандартное отклонение, минимальное и максимальное значения) для числовых признаков и уникальные значения для категориальных признаков в датасете.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Прочтение файла train
titanic_data = pd.read_csv('./files/train.csv')

# Количество признаков и их типы
print("Количество признаков и их типы:")
print(titanic_data.info())

# Проверка наличия пропущенных данных
print("\nПропущенные данные:")
print(titanic_data.isnull().sum())

# Основные статистические характеристики для числовых признаков
print("\nОсновные статистические характеристики для числовых признаков:")
print(titanic_data.describe())

# Уникальные значения для категориальных признаков
print("\nУникальные значения для категориальных признаков:")
print(titanic_data.select_dtypes(include=['object']).nunique())

print('Уникальные значения для столбцов Survived, Pclass')
print('Survived: ', titanic_data['Survived'].unique())
print('Pclass: ',titanic_data['Pclass'].unique())
print('Sex: ', titanic_data['Sex'].unique())
print('Embarked: ', titanic_data['Embarked'].unique())
```

Задание 2:
Построение графика "распределения выживших по полу" с использованием библиотеки matplotlib. Создайте столбчатую или круговую диаграмму, которая покажет, сколько мужчин и женщин выжило и погибло.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Прочтение файла train
titanic_data = pd.read_csv('./files/train.csv')

# Построение графика "распределения выживших по полу"
survived_sex = titanic_data.groupby(['Sex', 'Survived']).size().unstack()
survived_sex.plot(kind='bar', stacked=True)
plt.title("Распределение выживших по полу")
plt.xlabel("Пол")
plt.ylabel("Количество")
plt.xticks(rotation=0)
plt.legend(["Погибли", "Выжили"])
plt.show()

```

Задание 3:
Создание scatter plot для анализа взаимосвязи между возрастом и ценой билета. Используя библиотеку matplotlib, постройте точечную диаграмму, где по оси X будет возраст пассажира, а по оси Y - цена билета. Отобразите выживших и погибших пассажиров разными цветами или маркерами.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Прочтения файла train
titanic_data = pd.read_csv('./files/train.csv')

# Создание scatter plot для анализа взаимосвязи между возрастом и ценой билета
plt.scatter(titanic_data['Age'], titanic_data['Fare'], c=titanic_data['Survived'], cmap='viridis', alpha=0.5)
plt.title("Взаимосвязь между возрастом и ценой билета")
plt.xlabel("Возраст")
plt.ylabel("Цена билета")
plt.colorbar(label='Выжил')
plt.show()
```
